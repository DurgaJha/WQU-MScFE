```
import pandas_datareader.data as web
import datetime

# Define the start and end dates
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=365*10)

# Fetch the data from FRED
data = web.DataReader(["VXNCLS","RVXCLS"], 'fred', start_date, end_date)
#The first argument is the series ID, the second is data source and last two are the start and end date

# Print the data
print(data)
```

```
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['VXNCLS'], label="CBOE NASDAQ 100 Volatility Index")
plt.plot(data.index, data['RVXCLS'], label ="CBOE Russell 2000 Volatility Index")

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Volatility Index')
plt.title('Volatility Index (Last 10 Years)')
plt.legend()

# Customize the plot (optional)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
