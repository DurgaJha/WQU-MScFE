```python
import yfinance as yf
import matplotlib.pyplot as plt

data = yf.download(["AAPL", "TSLA", "AMZN"], start="2021-01-01", end="2023-12-31")["Adj Close"]
```

```python
data.head(5).round(2)
```

```python

# Plot the data with labels
plt.plot(data["AAPL"], label="AAPL")
plt.plot(data["TSLA"], label="TSLA")
plt.plot(data["AMZN"], label="AMZN")

# Add labels and legend
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price")
plt.legend()

# Show the plot
plt.show()

# Calculate and print the correlation matrix
print(data.corr())
```

```python
df = yf.download(["^IRX", "^FVX", "^TNX"], start="2021-01-01", end="2023-12-31")["Adj Close"]
df.head(5).round(4)
```
